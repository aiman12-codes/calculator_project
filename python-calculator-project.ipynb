**Introduction:**

In this notebook, I have created a simple calculator using Python.
It performs basic arithmetic operations and runs in a menu-based format where the user can repeatedly choose operations.

**Objective:**

*The main objective of this project is to understand and practice basic Python concepts such as:*

1.Functions

2.Loops

3.Conditional statements

4.User input handling

**Features of the Calculator:**

*This calculator can perform the following operations:*

1.Addition of two numbers

2.Subtraction of two numbers

3.Multiplication of two numbers

4.Division of two numbers (with error handling for division by zero).

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero"
    return a / b


while True:
    print("\n=== SIMPLE CALCULATOR ===")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "5":
        print("Program ended. Thank you!")
        break

    if choice in ["1", "2", "3", "4"]:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == "1":
            print("Result:", add(num1, num2))
        elif choice == "2":
            print("Result:", subtract(num1, num2))
        elif choice == "3":
            print("Result:", multiply(num1, num2))
        elif choice == "4":
            print("Result:", divide(num1, num2))
    else:
        print("Invalid choice. Please try again.")
```

**Working:**

*The program:*

1.Displays a menu of operations

2.Takes user input for choice

3.Accepts two numbers from the user

4.Displays the result

5.Repeats until the user chooses to exit
